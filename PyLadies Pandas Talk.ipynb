```python
complaints_by_zipcode_state = consumer_complaints.groupby(['zip_code', 'state'], as_index=True)['complaint_id'].count().order().reset_index()
complaints_by_zipcode_state.rename(columns={'complaint_id' : 'complaints_by_zip_and_state'}, inplace=True)
complaints_by_zipcode_state
```

```python
complaints_by_timely_state = consumer_complaints.groupby(['state', 'timely_response'], as_index=True)['complaint_id'].count().reset_index()
complaints_by_timely_state.rename(columns={'complaint_id' : 'complaints_by_state_and_timely_response'}, inplace=True)
```

```python
new_df = pandas.merge(complaints_by_zipcode_state, complaints_by_timely_state, on='state', how='outer')
```
